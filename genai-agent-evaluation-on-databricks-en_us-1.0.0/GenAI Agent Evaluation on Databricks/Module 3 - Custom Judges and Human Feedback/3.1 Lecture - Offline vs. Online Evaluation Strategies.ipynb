# Lecture - Offline vs. Online Evaluation Strategies

## Overview

Effective agent evaluation requires both offline validation using curated datasets and online monitoring of production performance. This lecture explores the complementary roles of offline and online evaluation, their respective strengths and limitations, and how to create feedback loops between them.

We'll examine when to use each approach, understand their implementation patterns, and learn how to build evaluation systems that evolve with real-world usage while maintaining rigorous quality standards.

**Learning Objectives**

By the end of this lecture, you will be able to:
- Identify when to use offline versus online evaluation strategies
- Understand the strengths and limitations of each approach
- Describe how offline and online evaluation complement each other
- Explain how to create feedback loops between production and evaluation
- Recognize best practices for comprehensive evaluation strategies

## A. Evaluation Strategy Overview

### A1. Two Complementary Approaches

![offline-vs-online-evaluation.png](../Includes/images/Evaluation with MLflow/offline-vs-online-evaluation.png "offline-vs-online-evaluation.png")

Agent evaluation requires both controlled validation and real-world monitoring. Offline and online evaluation serve different but complementary purposes in ensuring agent quality throughout the development and deployment lifecycle.

## B. Offline Evaluation: Pre-Deployment Validation

### B1. Controlled Environment Testing

Offline evaluation tests your agent using curated datasets before deployment. This approach provides controlled, reproducible validation of agent quality.

**Characteristics of offline evaluation:**

- **Controlled environment**: You control exactly what inputs the agent receives
- **Reproducibility**: Same dataset yields consistent evaluation results (modulo LLM non-determinism)
- **Ground truth availability**: You can include expected outputs, facts, or expert-labeled guidelines
- **Comprehensive coverage**: Intentionally test edge cases, failure modes, and diverse scenarios
- **Pre-deployment gate**: Prevents deploying agents that don't meet quality standards

### B2. Offline Evaluation Workflow

**Offline evaluation workflow:**

1. **Dataset creation**: Curate representative examples covering expected use cases, edge cases, and known failure patterns
2. **Ground truth labeling**: Add expected outputs, facts, or guidelines (often with domain expert involvement)
3. **Agent execution**: Run your agent on the dataset to generate responses
4. **Scorer application**: Apply multiple judges to assess various quality dimensions
5. **Analysis**: Review aggregated metrics and individual failures to identify improvements
6. **Iteration**: Update agent, re-evaluate, compare results

### B3. Offline Evaluation Strengths and Limitations

**Strengths:**
- Enables rigorous validation before users encounter your agent
- Supports A/B testing different agent configurations
- Provides baseline metrics for production monitoring

**Limitations:**
- Dataset may not fully represent real user behavior
- Static datasets become stale as usage patterns evolve
- Cannot capture issues that only emerge with scale or diverse users

## C. Online Evaluation: Production Monitoring

### C1. Real-World Performance Analysis

Online evaluation analyzes agent performance using real user interactions in production. This approach captures how your agent performs with actual usage patterns, unexpected inputs, and diverse user populations.

**Characteristics of online evaluation:**

- **Real-world data**: Evaluates against actual user queries and behaviors
- **Scale validation**: Tests how your agent handles production load and request diversity
- **Drift detection**: Identifies when performance degrades over time
- **Continuous monitoring**: Ongoing quality signals rather than one-time assessment
- **User feedback integration**: Combines automated evaluation with actual user satisfaction

### C2. Online Evaluation Workflow

**Online evaluation workflow:**

1. **Inference logging**: Deploy agent with AI Gateway-enhanced inference tables enabled
2. **Automatic tracing**: Every production request creates traces captured in inference tables
3. **Scheduled evaluation**: Periodically run scorers on recent production traces
4. **Alert configuration**: Set up monitoring alerts for quality degradation
5. **Feedback collection**: Gather user thumbs up/down or explicit feedback
6. **Dataset augmentation**: Extract interesting production examples for offline evaluation

### C3. MLflow's Production Monitoring Capabilities

**MLflow's production monitoring capabilities:**

Databricks provides built-in production monitoring for agents deployed via Model Serving:
- **Automated scorer execution**: Run MLflow scorers on production traces automatically
- **Quality dashboards**: Visualize evaluation metrics over time
- **Alerting**: Get notified when quality metrics drop below thresholds
- **Trace inspection**: Deep dive into individual failures or edge cases

### C4. Online Evaluation Strengths and Limitations

**Strengths:**
- Reflects actual user experience and real-world performance
- Detects issues that weren't anticipated in offline testing
- Provides data for continuous improvement

**Limitations:**
- Users may experience failures before you detect them
- More difficult to attribute causes (confounding from diverse inputs)
- Requires production infrastructure and monitoring systems

## D. Complementary Strategies

### D1. When to Use Each Approach

Effective agent evaluation combines both offline and online approaches:

**Offline evaluation for:**
- Pre-deployment validation and quality gates
- Comparing alternative agent implementations
- Testing specific hypotheses about agent behavior
- Rapid iteration during development

**Online evaluation for:**
- Validating that offline results generalize to production
- Detecting performance degradation or drift
- Understanding real user needs and pain points
- Building evaluation datasets from actual usage

### D2. Creating a Feedback Loop

**Creating a feedback loop:**

1. Start with offline evaluation on carefully curated datasets
2. Deploy to production with monitoring enabled
3. Analyze production traces to identify failures and edge cases
4. Add production examples to your offline evaluation dataset
5. Use enhanced dataset to validate improvements before re-deploying

This cycle ensures your evaluation evolves with your understanding of real-world usage while maintaining rigorous pre-deployment validation.

## E. Building Comprehensive Evaluation Datasets

### E1. Dataset Composition Principles

The quality of your evaluation depends critically on your dataset. A well-designed evaluation dataset should be representative, diverse, and include appropriate ground truth.

**Dataset composition principles:**

**1. Representativeness**: Include examples reflecting typical user queries and use cases
- Analyze production logs or user research to identify common patterns
- Ensure distribution of query types matches expected usage
- Include varying query complexity levels

**2. Edge case coverage**: Deliberately test boundary conditions and potential failures
- Ambiguous queries that could be interpreted multiple ways
- Queries requiring information not available in retrieval corpus
- Adversarial examples designed to trigger failure modes
- Queries in unexpected formats or phrasings

### E2. Diversity and Ground Truth Approaches

**3. Diversity dimensions**: Vary multiple aspects of inputs
- Query length (terse vs. verbose)
- Query complexity (simple fact lookup vs. multi-step reasoning)
- Domain coverage (different topics your agent should handle)
- User expertise levels (novice vs. expert domain knowledge)

**Ground truth approaches:**

**Expert labeling**: Domain experts create expected outputs or fact sets
- Most accurate but resource-intensive
- Essential for high-stakes applications
- Creates reusable gold standard dataset

**Synthetic generation**: Use LLMs to generate expected responses
- Scalable but requires validation
- Useful for bootstrapping evaluation
- Should be spot-checked by humans

**Production mining**: Extract examples from inference tables
- Reflects real usage patterns
- May require filtering or cleaning
- Can identify emergent failure modes

### E3. Dataset Maintenance

**Dataset maintenance:**

Evaluation datasets require ongoing curation:
- Regularly add new examples from production
- Remove or update stale examples
- Rebalance categories as usage patterns shift
- Version datasets alongside agent versions

## F. Integration with Unity Catalog

### F1. Governance Layer for Evaluation

Unity Catalog provides the governance layer for your evaluation workflow, ensuring traceability, access control, and lineage tracking.

**Agent registration**: When you register agents in Unity Catalog:
- All agent dependencies (vector indexes, UC functions, model endpoints) are tracked
- Access controls determine who can evaluate or modify agents
- Lineage shows which evaluation datasets and metrics produced each version
- Aliases (like `@champion`) enable clean promotion workflows

### F2. Dataset and Trace Storage

**Evaluation dataset storage**: Store datasets as Delta tables or volumes:
- Version datasets alongside agent versions
- Apply access controls to sensitive evaluation data
- Enable SQL-based dataset analysis and augmentation
- Support collaborative dataset curation

**Trace storage**: MLflow traces are stored in Unity Catalog:
- Query traces using SQL for analysis
- Join traces with evaluation results for deeper insights
- Apply data governance policies to production trace data
- Enable cross-team trace sharing (where appropriate)

This integration ensures your evaluation infrastructure benefits from the same governance, security, and collaboration capabilities as the rest of your data platform.

## G. Key Takeaways

Effective agent evaluation requires both offline and online strategies:

1. **Complementary approaches**: Offline evaluation provides controlled validation while online evaluation captures real-world performance
2. **Feedback loops**: Production data should inform offline evaluation datasets, creating continuous improvement cycles
3. **Dataset quality**: Representative, diverse, and well-maintained datasets are crucial for meaningful evaluation
4. **Governance integration**: Unity Catalog provides the infrastructure for scalable, governed evaluation workflows
5. **Continuous evolution**: Evaluation strategies must adapt as agents, usage patterns, and requirements evolve

The combination of rigorous offline validation and comprehensive online monitoring creates a robust evaluation framework that ensures agent quality throughout the development and deployment lifecycle.

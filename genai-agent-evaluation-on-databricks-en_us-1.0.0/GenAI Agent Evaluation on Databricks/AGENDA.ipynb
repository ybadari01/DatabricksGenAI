## GenAI Agent Evaluation on Databricks

This course teaches students how to systematically evaluate AI agents using MLflow's evaluation framework, addressing the unique challenges of non-deterministic AI systems that traditional software testing cannot handle. Students learn to implement various evaluation approaches including built-in judges for common criteria like correctness and safety, guideline judges for business-specific requirements, and custom judges for specialized needs. The course covers both offline evaluation using curated datasets and online production monitoring, with hands-on experience using MLflow's tracing capabilities to understand agent execution patterns and collect human feedback from different stakeholder types. Through practical demonstrations and labs, students develop skills in creating evaluation workflows that drive continuous quality improvements throughout the AI agent development lifecycle.

## Prerequisites

### Python-Specific Skills
- Basic Python syntax and data structures (lists, dictionaries)
- Understanding of functions, classes, and object-oriented programming concepts
- Experience with Python package management and imports
- Familiarity with JSON data handling and file operations
- Basic understanding of lambda functions and list comprehensions

### SQL-Specific Skills
- Basic SQL query syntax (SELECT, FROM, WHERE)
- Understanding of table joins and aggregations
- Knowledge of SQL functions and data types
- Experience with Unity Catalog SQL functions and procedures

### Databricks-Specific Skills
- Understanding of Databricks workspace navigation and notebook interface
- Knowledge of Unity Catalog structure (catalogs, schemas, tables, volumes)
- Experience with Databricks compute resources and serverless computing
- Familiarity with MLflow experiment tracking and model registry
- Understanding of Databricks model serving endpoints and deployment
- Awareness of knowledge of Delta tables

### GenAI/Agent-Specific Skills
- Basic understanding of Large Language Models (LLMs) and their capabilities
- Knowledge of prompt engineering and system prompts
- Understanding of tool-calling agents and function calling concepts
- Somewhat familiar with evaluation metrics for AI systems (correctness, safety, relevance)
- Basic knowledge of MLflow tracing and agent evaluation frameworks
- Somewhat familiar with the notion of human feedback collection and assessment workflows

### Other/Optional Skills
- Experience with YAML configuration files can be helpful
- Basic understanding of REST APIs and HTTP requests can be helpful
- Understanding of version control and model lifecycle management for ML models can be helpful

## Requirements

Please review the following requirements before starting the lesson:

* To run demo and lab notebooks, you need to use **Serverless compute (version 4)**, which is enabled by default.
* This course requires access to MLflow's evaluation framework and Unity Catalog for agent registration and governance.
* Some demonstrations require model serving endpoints for agent deployment and evaluation.

The course is structured into three logical modules:

1. **Module 1** introduces the fundamental challenges of AI agent evaluation and MLflow's framework
2. **Module 2** covers the core evaluation approaches using built-in and guideline judges
3. **Module 3** explores advanced topics including custom judges, human feedback, and evaluation strategies

Each module combines lectures for conceptual understanding with hands-on demos and labs for practical implementation. The progression moves from basic concepts to advanced techniques, ensuring students build a comprehensive understanding of AI agent evaluation using MLflow on Databricks.

```python
agent = demo_setup.load_agent("airbnb_eval_agent")
```

### Sample Questions

Now that we have our agent working, let's pass some questions using the `predict()` method defined in `airbnb_agent.py`. Let's first create a helper function we can use to send the expected payload to the agent.

```python
def agent_payload(question: str):
    return [{"input": [{"role": "user", "content": question}]}]
```

```python
agent.predict(agent_payload("What tools do you have available?"))
```

### Inspect the Trace

Based on the output above, we can see the tools the agent has access to. Let's next ask a question that will invoke a couple tool calls. Namely, the next question will nudge the agent to use both:

1. `avg_neigh_price`
2. `cnt_by_room_type`

The screenshot below is an example of the output when running the next cell and what you should expect. Notice the tools defined above are being called.

![mlflow-toolcall.png](../Includes/images/built-in agents with mlflow/mlflow-toolcall.png "mlflow-toolcall.png")

```python
agent.predict(agent_payload("Can you tell me what the average price is in Mission? Also, what's the number of listings for that neighborhood that have private rooms?"))
```

## Part 2 - Inspect the Trace in Unity Catalog

Navigate to the **Catalog Explorer** and search for **airbnb_eval_agent**. There you will find a model (at least **Version 1**) with the alias **@champion**. Click on it and navigate to the **Traces** tab. There you will find the same two requests we sent above.

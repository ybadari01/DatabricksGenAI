# Lab - Building a Single Agent with Agent Bricks: Supervisor Agent

## Overview

Agent Bricks Supervisor Agent creates a coordinated multi-agent system that orchestrates AI agents and tools to work together on complex tasks. This system can coordinate Genie Spaces, agent endpoints, Unity Catalog functions, and MCP servers to deliver comprehensive solutions across specialized domains.

The supervisor system uses advanced AI orchestration patterns to manage agent interactions, task delegation, and result synthesis. You can improve coordination quality over time with natural language feedback from subject matter experts.

- Clean up deployed Agent Bricks resources

## C. Configuring the Supervisor Agent

Now that you have Unity Catalog functions ready, you can create and configure the supervisor agent system.

### C1. Access Agent Bricks Interface

1. Navigate to **Agents** in the left navigation pane of your workspace.
2. Click on **Create Agent** and select **Supervisor Agent**.
3. You'll be directed to the configuration interface.

### C2. Configure Supervisor Settings

On the **Configure** tab, set up your supervisor with the following information:

- **Name**: Name your agent your lab username + MAS_single_agent. For example, `labuser_123_abc_MAS_single_agent`
- **Description**: Copy and paste the following description: _This is an agent that uses tool calling to answer questions about the Airbnb sample dataset._

### C3. Add Unity Catalog Functions as Tools

Under **Configure Agents**, add your Unity Catalog functions:

1. Click **+ Add** to add a new agent/tool
1. In the **Type** field, select **Unity Catalog Function**
1. Select the following functions from the **Unity Catalog Function** drop-down menu (you can use the search box)
   - `avg_neigh_price`
1. Click **Confirm** and repeat the same instructions to add the UC function `airbnb_posting_info`
1. Provide an **Agent Name** for this tool or you can use the automatically generated Agent Name provided after selecting the tool.
1. Typically, under **Describe the content**, you should provide a detailed description of what this function does and when it should be used. However, because we have already pre-configured our function with a description, that has been brought in from UC.

**NOTE:** You can add up to 10 agents/tools to a single supervisor system.

### C4. Complete Supervisor Creation

After configuring all settings and adding your tools:

1. Review your configuration for accuracy
2. Click **Create Agent**
3. Wait for the system to build your supervisor agent (this will take a few minutes)

## E. (Optional) Implementing Feedback and Improvement

Agent Bricks Supervisor Agent can improve coordination quality based on natural language feedback from subject matter experts.

### E1. Create Labeling Sessions
_It usually takes approximately 15 minutes for the agent to be ready for a SME session starting from the moment its endpoint has been created._

To gather feedback for supervisor improvement:

1. Navigate to the **Examples** tab back in your Agent's configuration menu
2. Click **+ Add** to add task scenarios for evaluation
3. Enter questions or tasks that test supervisor coordination, e.g. _What is the average price for Upper Market?_
4. Click **Add**

### E2. Mock Expert Feedback

Set up a labeling session for subject matter experts:

1. After clicking on the question, you will see  **Input** and **Guidlines**. Click on **Guidelines**.
2. Add in the guideline: _Upper Market is also called "Castro/Upper Market" so you should use that in your query._
3. Click **Save**

### E3. Apply Feedback and Retrain

After experts complete their reviews:

1. Return to the **Build** tab
2. Test the supervisor again to validate improvements with the same question that you passed earlier: _What is the average price for Upper Market?_

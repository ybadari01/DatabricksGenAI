# Demo - Building a Knowledge Assistant (KA) Agent with Agent Bricks

## Overview

In this demo, we’ll explore how to build a high-quality knowledge assistant agent using **Agent Bricks: Knowledge Assistant**. We’ll walk through creating a question-and-answer chatbot over your documents using Databricks’ declarative agent creation tool, and then improve its quality through expert feedback and labeling sessions.

**Scenario:** The **Orion Knowledge Assistant (OKA)** empowers engineers and technicians working on the Orion A1 humanoid platform by delivering instant, context-grounded answers sourced from internal design manuals, compliance documents, and maintenance guides stored in Databricks. When a field engineer asks how to recalibrate a motion controller or verify a firmware checksum, OKA retrieves the exact procedure and references the correct section—or clearly states when the information isn't available. This approach maintains trust and accuracy in mission-critical environments.

   
## A. Creating the KA Agent

**Agent Name:** **Orion Knowledge Assistant (OKA)**

The documents are placed in a Unity Catalog (UC) volume. We'll create a KA agent using Agent Bricks.

**Steps to create your agent:**

1. In your Databricks workspace, navigate to **Agents** in the left navigation pane
1. Click **Create Agent** button
1. Click on the **Knowledge Assistant** box to start creating a new KA agent
1. Provide the following information:
   - **Name**: `Orion_Knowledge_Assistant`
   - **Description**: Enter a description such as:  
     `Orion Knowledge Assistant (OKA) helps engineers and technicians quickly find accurate answers from Orion's internal manuals, maintenance guides, and safety documents. It delivers clear, verified responses with source references, reducing search time and ensuring consistent, reliable information across teams.`
   - **Knowledge source type**: Files in a Volume
   - **Source**: Select the Unity Catalog volume based on the catalog you used (see above for the volume name)
   - Click **Confirm**
   - **Knowledge source name**: `Company Documents`
   - **Content description**: Enter a content description such as:  
     `Contains Orion technical documentation, engineering notes, handbooks, and frequently asked questions.`
1. *(Optional)* Add instructions for how the agent should respond. See example instructions below.
1. Click **Create Agent** to start the creation process

Sample instruction prompt:  
`You are the Orion Knowledge Assistant (OKA). Respond in a clear, professional, and factual tone appropriate for engineers and technical staff. Use only verified information from Orion's internal documents, and include source references when available. If the answer cannot be found, clearly state that and suggest related sections or next steps. Do not speculate, make assumptions, or provide information outside the provided context.`

**⏳ Note:** It can take up to 10 minutes to create your agent and sync the knowledge sources.

## B. Testing Your Agent

After your agent has finished building, you can test its functionality using the built-in chat interface. The chat area appears on the right side of the Agent Bricks interface.

**Sample questions to test:**

Try asking your agent these questions based on the sample documents:

1. "How does Orion verify compliance with ISO 13849-1?"
1. "How does the Orion motion controller maintain stability during high-speed movement?"
1. **"What does the red blinking light on Orion mean?"** 
*Note: There is no red blinking light in Orion documentation.* 

**What to look for:**
- Accurate answers based on your documents
- Proper citations and source references
- Appropriate handling of questions outside the knowledge base
- Professional and helpful tone

**💡 If you want the agent to provide a better answer for the last question, you will need to improve agent quality. Let's move to that next!**

   
## C. Improving Agent Quality

Agent Bricks: Knowledge Assistant enables you to enhance your agent's quality through expert feedback and labeling sessions. These features allow you to collect natural language feedback from subject matter experts and use it to retrain and optimize your agent's performance.

**When to use quality improvement:**
- When your agent provides inaccurate or incomplete responses
- To fine-tune the agent's tone and communication style
- When expanding to new knowledge domains or use cases
- For ongoing optimization based on user feedback
- To ensure consistent performance across different question types





### C1. Using Labeled Data for Quality Improvement

In this step, we will provide labeled data to improve the agent's response to the question above.

**Quality improvement process:**
1. Navigate to the **Examples** tab in your agent's interface.
1. Click the **Add** button.
1. Enter the question (`What does the red blinking light on Orion mean?`) and press **Add** to add the question.
1. Click the question to open its details.
1. Add **Guidelines** such as:
  
    >Inform the user that there isn't a blinking red light on Orion

    >Ask the user to restart Orion by removing and reinserting the battery
      
    >Check the light color
      
    >Check the light again to see if it blinks
  
1. Click on **Save**.

Add evaluation questions that test different aspects of your agent's knowledge.

**Test the Agent:**
1. Return to the agent and ask the same question to **verify if the response has improved**. You can also change your question to **ask about another color to see how the answer adapts**. 

### C2. Using Expert Review for Quality Improvement

**Labeling sessions and expert review:**

The labeling session feature allows experts to:
- Review agent responses to evaluation questions
- Provide natural language feedback on response quality
- Add guidelines and expectations for agent behavior
- Evaluate responses for accuracy, completeness, and tone

For step-by-step instructions on the expert review process, see the [Agent Bricks: Knowledge Assistant documentation – Step 3: Improve quality](https://docs.databricks.com/aws/en/generative-ai/agent-bricks/knowledge-assistant#step-3-improve-quality).

**Best Practice:** Quality improvement is an iterative process. Plan for multiple rounds of feedback collection and refinement to achieve optimal agent performance.

## D. Cleanup Resources

After completing this demo, it's important to clean up the resources you created to avoid incurring unnecessary costs. **The KA agent creates a vector search endpoint and a serving endpoint in the background** to enable semantic search over your documents. **To delete your agent:**
1. Navigate to **Agents** in the left navigation pane
1. Locate your **Orion_Knowledge_Assistant** agent
1. Click on the agent to open its details

Deleting the agent will also remove the associated vector search endpoint and serving endpoint.
